```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px

data = pd.read_csv("./GvpElg_ie7ekdOI_e3GSl4nV9wsvUCXOoQ-jIkDhGzUUh1hiRKVu!-1610761495!-2050938374!1641828192226 (2).csv", skiprows=1)

#data = pd.read_csv("./GvpElg_ie7ekdOI_e3GSl4nV9wsvUCXOoQ-jIkDhGzUUh1hiRKVu!-1610761495!-2050938374!1641828192226 (1).csv")
df = pd.DataFrame(data)
conditions = [
    (df['Area (Sqft)'] < 500),
    (df['Area (Sqft)'] < 650),
    (df['Area (Sqft)'] < 800),
    (df['Area (Sqft)'] < 1000),
    (df['Area (Sqft)'] > 1000),
    ]
df['size'] = np.select(conditions, ['studio', '2 bed room', '3 bed room', 'large',  'super large'])
df['Date of Sale'] = pd.to_datetime(df['Date of Sale'])
studio_sqdf = df[df['size'].str.contains('studio')].groupby(['Project Name', 'Date of Sale'])['Unit Price ($psf)'].mean().reset_index(name="mean").sort_values('Date of Sale', ascending=False)
twobed_sqdf = df[df['size'].str.contains('2 bed')].groupby(['Project Name', 'Date of Sale'])['Unit Price ($psf)'].mean().reset_index(name="mean").sort_values('Date of Sale', ascending=False)
ndf =df.groupby(['size', 'Date of Sale'])['size'].count().reset_index(name="count").sort_values('Date of Sale', ascending=False)
```

```python
fig = px.line(ndf, x="Date of Sale", y="count", color='size', title='Transactions based on the property size')
fig.show()
```

```python
fig = px.line(studio_sqdf, x="Date of Sale", y="mean", color='Project Name', title='Studio sqft unit price per project'))
fig.show()
```

```python
fig = px.line(twobed_sqdf, x="Date of Sale", y="mean", color='Project Name', title='2 bedroom sqft unit price per project)
fig.show()
```
